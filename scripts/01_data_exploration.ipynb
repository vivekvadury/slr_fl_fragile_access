```python
public_schools = gpd.read_file(r"../data/raw/services/primary_schools/EDGE_GEOCODE_PUBLICSCH_2324.shp")
private_schools = gpd.read_file(r"../data/raw/services/primary_schools/EDGE_GEOCODE_PRIVATESCH_2324.shp")
```
